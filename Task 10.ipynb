#### Task 13: Simulate rolling a six-sided die 1000 times and count the frequency of each outcome

```python
die_rolls = np.random.randint(1, 7, size=1000)
outcome_counts = np.bincount(die_rolls)[1:7]  # We use [1:7] to exclude the zero-count for value 0 which is not a die roll outcome
print("Frequency of each outcome from rolling a six-sided die 1000 times:", outcome_counts)
```

#### Q2: In a normal distribution, 68% of the data falls within one standard deviation of the mean. If a dataset has a mean of 50 and a standard deviation of 5, between what two values does 68% of the data lie?

```python
mean = 50
std_dev = 5

lower_bound = mean - std_dev
upper_bound = mean + std_dev

print("lower bound:",lower_bound)
print("upper bound:",upper_bound)
```

#### Q3: A fair coin is flipped 10 times. What is the probability of getting exactly 6 heads?

```python
import numpy as np
from math import comb

n = 10
p = 0.5
k = 6

probability = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

probability
```
