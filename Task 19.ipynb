```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```


#### Linear Regression Plot (sns.regplot)

```python
x = [5, 15, 25, 35, 45, 55]
y = [5, 20, 14, 32, 22, 38]

sns.regplot(x=x, y=y)
plt.title('Linear Regression Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### Multiple Linear Regression Plot (sns.lmplot)

```python
data = pd.DataFrame({
    'x1': [1, 2, 3, 4, 5],
    'x2': [10, 20, 30, 40, 50],
    'y': [5, 15, 25, 35, 45]
})
sns.lmplot(x='x1', y='y', data=data, height=6, aspect=1.5)
plt.title('Multiple Linear Regression Plot')
plt.xlabel('X1')
#Multiple Predictors
sns.lmplot(x='x2', y='y', data=data)
plt.ylabel('Y')
plt.show()
```

#### Residentail Plots

```python
np.random.seed(0)
x = np.random.rand(100) * 10
y = 2.5 * x + np.random.randn(100) * 2  # True relationship with noise

# Fit a linear regression model
sns.regplot(x=x, y=y, ci=None, line_kws={"color": "red"})  
sns.residplot(x=x, y=y, color='blue')

plt.title('Residual Plot')
plt.xlabel('X')
plt.ylabel('Residuals')
#Lowess Smoothing
sns.residplot(x=x, y=y, lowess=True, color='green')
plt.show()
```

#### Histogram (sns.histplot)


```python
np.random.seed(0)
data = np.random.randn(1000)  

# Create histogram with seaborn
sns.histplot(data, bins=30, color='blue')

plt.xlabel('Values')
plt.ylabel('Frequency')
##sns.histplot(data, bins=30, kde=True, color='blue')
plt.show()
```

#### Kernel Density Estimate (KDE) Plot (sns.kdeplot)

```python
np.random.seed(0)
data = np.random.randn(1000) 
sns.kdeplot(data, fill=True, color='blue')

plt.title('Kernel Density Estimate (KDE) Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
#Adjusting Bandwidth:
sns.kdeplot(data, fill=True, bw_method=0.5, color='green')
plt.xlabel('Values')
plt.show()
```

#### Distribution Plot (sns.distplot)

```python
np.random.seed(0)
data = np.random.randn(1000)  

# Create distribution plot with seaborn
sns.distplot(data, bins=30, hist=True, kde=True, color='blue',
             hist_kws={'edgecolor': 'black'}, kde_kws={'linewidth': 2})

plt.title('Distribution Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

#### Box Plot (sns.boxplot)

```python
np.random.seed(0)
data = np.random.randn(100, 4)  

sns.boxplot(data=data, palette='Set3')

plt.title('Box Plot using Seaborn')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

#### Violin Plot (sns.violinplot)

```python
np.random.seed(0)
data = np.random.randn(100, 4)  # Sample data with 4 columns and 100 rows

# Create violin plot with seaborn
sns.violinplot(data=data, palette='muted')

plt.title('Violin Plot ')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

#### Pair Plot (sns.pairplot)


```python
np.random.seed(0)
data = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])

# Create pair plot with seaborn
sns.pairplot(data)
plt.suptitle('Pair Plot', y=1.02) 
plt.show()
```

#### Heatmap (sns.heatmap)

```python
np.random.seed(0)
data = np.random.rand(10, 12)  
# Create heatmap with seaborn
sns.heatmap(data, cmap='viridis')

plt.title('Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Adding annotations
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt='.2f')
plt.show()
```

```python
#Adjusting colorbar
sns.heatmap(data, cmap='coolwarm', cbar=True, cbar_kws={'orientation': 'horizontal'})
plt.show()
```

#### Joint Plot (sns.jointplot)

```python
np.random.seed(0)
x = np.random.randn(1000)  
y = 2 * x + np.random.randn(1000)  

# Create joint plot with seaborn
sns.jointplot(x=x, y=y, kind='scatter', color='blue') #we can use kind as hex, kde, reg

plt.title('Joint Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### Facet Grid (sns.FacetGrid)

```python
np.random.seed(0)
data = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

# Create Facet Grid with seaborn
g = sns.FacetGrid(data, col='Category', hue='Category')
g.map(sns.scatterplot, 'X', 'Y')
g.add_legend()

plt.suptitle('Facet Grid', y=1.02)  # Adjust title position
plt.show()
```
