#### Group the DataFrame by a specific column and calculate the mean of each group.
Using groupby method to group a specific column and mean for calculating the mean

```python
df['new_strength'] = df['Strength'].str.replace(' mg', '').astype(float)
mean_strength_by_category = df.groupby('Category')['new_strength'].mean()
mean_strength_by_category
```

```python
df
```

```python
group2= df.groupby(['Category','Name'])['new_strength'].sum()
group2
```

#### Use the agg method to apply multiple aggregation functions to grouped data.
The agg() method in Pandas allows you to apply multiple aggregation functions to grouped data simultaneously. This is particularly useful when you want to calculate different statistics (like mean, sum, count, etc.) for specific columns within each group.

```python
agg_data=df.groupby(["Category", 'Name']).agg({'new_strength': ["mean","sum"]})
agg_data
```

#### Calculate the size of each group
We will use .size() method

```python
group_size=df.groupby(['Name', 'Category',"Manufacturer"]).size()
group_size
```

```python
group_size2=df.groupby(['Name', 'Category']).size()
group_size2
```

#### Use isin to filter rows based on a list of values.
