#### Task 1: Basic Plot

```python
#Simple plot with multiple lines and adding colours

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
#Adding Markers
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]


plt.plot(x, y, marker='p')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

plt.show()


#### Task 2: Scattered Plots

```python
#Creating a basic one first
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Creating one with styles and markers 
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.scatter(x, y, color='red', marker='^', s=100)  
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Multiple plots on the same graph
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

plt.scatter(x1, y1, color='purple', marker='o', s=100, label='Red Circle')
plt.scatter(x2, y2, color='blue', marker='>', s=100, label='Blue triangle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
#Adding titles and labels 
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='green', marker='s', s=100)
plt.title('Scatter Plot with Labels and Titles')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]},{y[i]})')

plt.show()
```

#### Task 3: Bar Charts

```python
#Creating a vertical bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
plt.bar(categories, values)
plt.title('Vertical Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Horizontal Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
plt.barh(categories, values)
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python

# Customized Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Customized Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python


# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

bars = plt.bar(categories, values, color=['skyblue', 'salmon', 'limegreen', 'gold', 'purple'], edgecolor='black', hatch='/')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # va: vertical alignment

plt.title('Vertical Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
bars = plt.barh(categories, values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], edgecolor='black')
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + 0.1, bar.get_y() + bar.get_height()/2.0, f'{xval:.2f}', ha='center', va='center')

plt.title('Horizontal Bar Chart with More Customizations')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add vertical gridlines
plt.show()
```

#### Task 4: Pie Charts

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]


plt.pie(sizes, labels=labels)
plt.show()


```

```python
#Customize Slice Colors and adding a legend
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['skyblue', 'salmon', 'limegreen', 'gold']
plt.pie(sizes, labels=labels, colors=colors)
plt.title(' Pie Chart with Legend')
plt.legend(labels, loc='best')
plt.show()
```

```python

# Pie Chart with Exploded Slice
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['skyblue', 'salmon', 'limegreen', 'gold']
explode = (0.1, 0.1, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, explode=explode)

plt.title('Pie Chart with Exploded Slice')
plt.show()
```

```python

# Pie Chart with Percentage Labels
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['skyblue', 'salmon', 'limegreen', 'gold']
explode = (0, 0.1, 0, 0)  

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.show()
```
