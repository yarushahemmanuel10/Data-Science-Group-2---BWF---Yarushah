## Task 1: Subplots:
* Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).
* Customize the subplots with titles, axis labels, and legends.


```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Generate some sample data

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
data = np.random.normal(0, 1, 1000)
```

### Create a figure with 2x2 subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
```

#### Accessing it now

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
#Subplot 1: Line Plot
axs[0, 0].plot(x, y1, label='sin(x)', color='blue')
axs[0, 0].plot(x, y2, label='cos(x)', color='red', linestyle='--')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()
# Subplot 2: Scatter Plot
axs[0, 1].scatter(x, y1, label='sin(x)', color='green', marker='o')
axs[0, 1].scatter(x, y2, label='cos(x)', color='purple', marker='x')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()
# Subplot 3: Bar Plot
axs[1, 0].bar(categories, values, color=['skyblue', 'salmon', 'limegreen', 'gold'])
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Subplot 4: Histogram
axs[1, 1].hist(data, bins=30, edgecolor='black', alpha=0.7)
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Task 2: Histograms



#### Generate sample data

```python
data1 = np.random.normal(0, 1, 1000)  # Normal distribution
data2 = np.random.normal(2, 1.5, 1000) 
```

```python
# Plotting histograms
plt.figure(figsize=(10, 6))

# Histogram for data1
plt.hist(data1, bins=30, alpha=0.5, label='Data 1', density=True)

# Histogram for data2
plt.hist(data2, bins=30, alpha=0.5, label='Data 2', density=True)

# Adding labels and titles
plt.title('Histogram Comparison of Data 1 and Data 2')
plt.xlabel('Value')
plt.ylabel('Normalized Frequency')
plt.legend()

plt.show()
```

## Task 3: Box Plots

#### Generate sample data

```python
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(2, 1.5, 100)
data3 = np.random.normal(-1, 1, 100)
```

#### Combine data into a list

```python
data = [data1, data2, data3]
```

```python
data = [data1, data2, data3]
# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['Data 1', 'Data 2', 'Data 3'], patch_artist=True)

# Customize box plot colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(plt.gca()['boxes'], colors):
    patch.set_facecolor(color)

# Adding labels and titles
plt.title('Box Plot Comparison of Data 1, Data 2, and Data 3')
plt.xlabel('Data Sets')
plt.ylabel('Values')

plt.show()
```
