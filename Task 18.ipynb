## Introduction to Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

### Scatter Plot
Scatter plots are useful for visualizing the relationship between two numerical variables.

```python
tips = sns.load_dataset("tips")

# Scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

### Line Plot
Line plots are suitable for showing trends over time or continuous variables.

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

### Bar Plot
Bar plots are used to compare categorical variables.

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Count Plot
Count plots show the counts of observations in each categorical bin using bars.

```python
sns.countplot(x="day", data=tips)
plt.title('Count Plot of Days')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

### Box Plot
Box plots show the distribution of categorical data and highlight outliers.

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

# Customization
#### Setting the Style
Seaborn has different built-in styles that can be set using sns.set_style().

```python
sns.set_style("whitegrid")

# Example plot with the new style
sns.barplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set3")
plt.title('Bar Plot with Whitegrid Style')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Changing Palette
Palettes can be changed using the palette parameter in Seaborn functions.

```python
sns.barplot(x="day", y="total_bill", data=tips, palette="husl")
plt.title('Bar Plot with "husl" Palette')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```
